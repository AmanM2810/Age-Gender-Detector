For Blurring images blur function is used

```python
img7=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
figure_size=7
img8=cv2.blur(img,(figure_size,figure_size))
plt.figure(figsize=(11,6))
plt.subplot(121),plt.imshow(img),plt.title("orignal")
plt.subplot(122),plt.imshow(img8),plt.title("blurred")
```

Zooming Image

```python
(height,width)=img2.shape[0:2]
zoom=cv2.resize(img2,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
zoom2 = cv2.resize(img2,(int(width*1.5),int(height*1.5)),interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom2)
```

Rotation

```python
m = cv2.getRotationMatrix2D((width/2,height/2),10,1)
rotate=cv2.warpAffine(img2,m,(width,height))
plt.imshow(rotate)
```

Translate 

```python
m = np.float32([[1,0,50],[0,1,100]])
translate = cv2.warpAffine(img2,m,(width,height))
plt.imshow(translate)
```
